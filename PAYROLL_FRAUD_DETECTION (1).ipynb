```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.cluster import KMeans

# Assume `data_cleaned` is already defined and cleaned

# Select relevant columns for anomaly detection
features = ['Base Pay', 'Bureau', 'Office', 'Job Code', 'Position ID', 'Fiscal Year']

# Create Isolation Forest model with selected features
model = IsolationForest(contamination=0.01, random_state=42)

# Fit the model and predict anomalies using the selected features
data_cleaned['anomaly'] = model.fit_predict(data_cleaned[features])
data_cleaned.loc[:, 'anomaly'] = np.where(data_cleaned['anomaly'] == -1, 1, 0)

# Risk scoring for detected anomalies
data_cleaned['risk_score'] = data_cleaned['anomaly'] * data_cleaned['Base Pay']  # Using loc to avoid SettingWithCopyWarning

# Save the updated dataframe to an Excel file
output_file = 'anomaly_detection_results_extended.xlsx'
data_cleaned.to_excel(output_file, index=False)

print(f"\nUpdated Excel file with anomalies and predictions saved to: {output_file}")

# Print updated records with anomaly column
print("\nUpdated Records with Anomalies:")
print(data_cleaned[['Fiscal Year', 'Fiscal Quarter', 'Fiscal Period', 'First Name', 'Last Name',
                    'Middle Init', 'Bureau', 'Office', 'Office Name', 'Job Code', 'Job Title',
                    'Base Pay', 'Position ID', 'Employee Identifier', 'Original Hire Date',
                    'anomaly', 'risk_score']].head())

# Print detected anomalies
anomalies = data_cleaned[data_cleaned['anomaly'] == 1]
print("\nRecords of Detected Anomalies:")
print(anomalies[['Fiscal Year', 'Fiscal Quarter', 'Fiscal Period', 'First Name', 'Last Name',
                 'Middle Init', 'Bureau', 'Office', 'Office Name', 'Job Code', 'Job Title',
                 'Base Pay', 'Position ID', 'Employee Identifier', 'Original Hire Date',
                 'anomaly', 'risk_score']].head())

# Visualize the anomalies
plt.figure(figsize=(12, 6))
sns.countplot(data=data_cleaned, x='anomaly')
plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel('Anomaly')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Normal', 'Anomaly'])
plt.show()

# Identify potential fraud payments (anomalies)
fraud_payments = data_cleaned[data_cleaned['anomaly'] == 1]
print("\nPotential Fraud Payments:")
print(fraud_payments[['Base Pay', 'anomaly']].head())

# Detect duplicate payments
duplicate_payments = data_cleaned[data_cleaned.duplicated(subset=['Base Pay'], keep=False)]
print("\nDuplicate Payments:")
print(duplicate_payments[['Base Pay', 'anomaly']])

# Risk scoring for anomalies
print("\nRisk Scoring for Detected Anomalies:")
risk_scoring_anomalies = anomalies[['Base Pay', 'risk_score']]
print(risk_scoring_anomalies)

# Summary statistics for detected anomalies
summary_stats = anomalies['Base Pay'].describe()
print("\nSummary Statistics for Detected Anomalies:")
print(summary_stats)

# High-Risk Detected Anomalies
high_risk_anomalies = anomalies[['Base Pay', 'risk_score']]
print("\nHigh-Risk Detected Anomalies:")
print(high_risk_anomalies)

# Check for class imbalance
print("\nClass Distribution:")
print(data_cleaned['anomaly'].value_counts())

# Prepare data for SMOTE
X = data_cleaned[features]
y = data_cleaned['anomaly']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Fit the model on resampled data
model_resampled = IsolationForest(contamination=0.01, random_state=42)
model_resampled.fit(X_resampled)

# Predict on test set
y_pred = model_resampled.predict(X_test)
y_pred = np.where(y_pred == -1, 1, 0)

# Calculate accuracy
accuracy = (y_test == y_pred).mean()
print(f"\nAccuracy: {accuracy:.2f}")

# Evaluation of the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import LocalOutlierFactor

# Load the dataset
df = pd.read_csv("/content/Employee_Payroll (1).csv")

# Fill missing values
for col in ['Base Pay', 'Fiscal Year']:
    df[col].fillna(df[col].mean(), inplace=True)  # Fill numerical columns with mean

for col in ['Job Title', 'Job Code', 'Position ID']:
    df[col].fillna(df[col].mode()[0], inplace=True)  # Fill categorical columns with mode


# Convert categorical columns to numerical values
label_encoders = {}
for col in ['Job Title', 'Job Code', 'Position ID']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Select relevant features
features = ['Base Pay', 'Fiscal Year', 'Job Title', 'Job Code', 'Position ID']
X = df[features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit the Local Outlier Factor model
lof = LocalOutlierFactor(n_neighbors=20, novelty=True)
lof.fit(X_scaled)

def predict_fraud(base_pay, fiscal_year, job_title, job_code, position_id):
    # Convert categorical inputs to numerical using label encoders
    job_title_encoded = label_encoders['Job Title'].transform([job_title])[0]
    job_code_encoded = label_encoders['Job Code'].transform([job_code])[0]
    position_id_encoded = label_encoders['Position ID'].transform([position_id])[0]

    # Prepare user input
    user_data = [[base_pay, fiscal_year, job_title_encoded, job_code_encoded, position_id_encoded]]

    # Scale the input data
    user_data_scaled = scaler.transform(user_data)

    # Predict with the model
    prediction = lof.predict(user_data_scaled)
    if prediction[0] == -1:
        return "Fraudulent"
    else:
        return "Non-fraudulent"

# Get user inputs
base_pay = float(input("Enter Base Pay: "))
fiscal_year = int(input("Enter Fiscal Year: "))
job_title = input("Enter Job Title: ")
job_code = input("Enter Job Code: ")
position_id = input("Enter Position ID: ")

# Run the prediction function
result = predict_fraud(base_pay, fiscal_year, job_title, job_code, position_id)
print("Fraud Prediction:", result)
```

The Local Outlier Factor (LOF) model gives an output based on detecting how isolated a given data point is from others in the dataset. Here's how it decides whether an input is "fraudulent" or "not fraudulent":

Distance from Nearest Neighbors: The LOF algorithm compares the distance of a data point to its nearest neighbors. Points that are far from their neighbors are flagged as anomalies, meaning they could be "fraudulent."

Density: LOF calculates the "local density" of each point and compares it to the densities of its neighbors. If a point has significantly lower density (indicating it’s in a sparse area), it's more likely to be labeled as "fraudulent."

Threshold: When you trained the model, a threshold was determined for what constitutes an "outlier" (potential fraud). If a new input falls outside this threshold, it's flagged as "fraudulent."

# Ensemble model (LOF AND ISOLATION FOREST) with grid search hyperparameter turning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")

# Select relevant columns for anomaly detection
features = ['Base Pay', 'Bureau', 'Office', 'Job Code', 'Position ID', 'Fiscal Year']

# Create a function for combining LOF and Isolation Forest
def ensemble_anomaly_detection(data, features):
    # Fit Isolation Forest
    isolation_forest = IsolationForest(contamination=0.01, random_state=42)
    data['isolation_anomaly'] = isolation_forest.fit_predict(data[features])
    data['isolation_anomaly'] = np.where(data['isolation_anomaly'] == -1, 1, 0)

    # Fit Local Outlier Factor
    lof = LocalOutlierFactor(n_neighbors=20)
    data['lof_anomaly'] = lof.fit_predict(data[features])
    data['lof_anomaly'] = np.where(data['lof_anomaly'] == -1, 1, 0)

    # Combine both models (logical OR)
    data['combined_anomaly'] = np.maximum(data['isolation_anomaly'], data['lof_anomaly'])

    return data

# Perform anomaly detection
data_cleaned = ensemble_anomaly_detection(data_cleaned, features)

# Risk scoring for detected anomalies
data_cleaned['risk_score'] = data_cleaned['combined_anomaly'] * data_cleaned['Base Pay']

# Save the updated dataframe to an Excel file
output_file = 'anomaly_detection_results_combined.xlsx'
data_cleaned.to_excel(output_file, index=False)
print(f"\nUpdated Excel file with combined anomalies and predictions saved to: {output_file}")

# Print updated records with combined anomaly column
print("\nUpdated Records with Combined Anomalies:")
print(data_cleaned[['Fiscal Year', 'Fiscal Quarter', 'Fiscal Period', 'First Name', 'Last Name',
                    'Middle Init', 'Bureau', 'Office', 'Office Name', 'Job Code', 'Job Title',
                    'Base Pay', 'Position ID', 'Employee Identifier', 'Original Hire Date',
                    'combined_anomaly', 'risk_score']].head())

# Print detected anomalies
anomalies = data_cleaned[data_cleaned['combined_anomaly'] == 1]
print("\nRecords of Detected Anomalies:")
print(anomalies[['Fiscal Year', 'Fiscal Quarter', 'Fiscal Period', 'First Name', 'Last Name',
                 'Middle Init', 'Bureau', 'Office', 'Office Name', 'Job Code', 'Job Title',
                 'Base Pay', 'Position ID', 'Employee Identifier', 'Original Hire Date',
                 'combined_anomaly', 'risk_score']].head())

# Summary statistics for detected anomalies
summary_stats = anomalies['Base Pay'].describe()
print("\nSummary Statistics for Detected Anomalies:")
print(summary_stats)

# High-Risk Detected Anomalies
high_risk_anomalies = anomalies[['Base Pay', 'risk_score']]
print("\nHigh-Risk Detected Anomalies:")
print(high_risk_anomalies)

# Check for class imbalance
print("\nClass Distribution:")
print(data_cleaned['combined_anomaly'].value_counts())

# Prepare data for SMOTE
X = data_cleaned[features]
y = data_cleaned['combined_anomaly']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Define the parameter grid for Isolation Forest
param_grid = {
    'contamination': [0.01, 0.05, 0.1],
    'max_samples': [0.5, 0.75, 1.0]
}

# Perform hyperparameter tuning for Isolation Forest
grid_search = GridSearchCV(IsolationForest(random_state=42), param_grid, cv=5, scoring='f1')
grid_search.fit(X_resampled)

# Best Isolation Forest model
best_isolation_forest = grid_search.best_estimator_
print(f"\nBest Isolation Forest Parameters: {grid_search.best_params_}")

# Fit the best model on resampled data
best_isolation_forest.fit(X_resampled)

# Predict on test set
y_pred = best_isolation_forest.predict(X_test)
y_pred = np.where(y_pred == -1, 1, 0)

# Calculate accuracy
accuracy = (y_test == y_pred).mean()
print(f"\nAccuracy: {accuracy:.2f}")

# Evaluation of the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Generate predictions from both models for the entire dataset
data_cleaned['isolation_pred'] = np.where(best_isolation_forest.predict(X) == -1, 1, 0)
data_cleaned['lof_pred'] = np.where(anomalies['lof_anomaly'].reindex(data_cleaned.index).fillna(0) == 1, 1, 0)

# Combine both predictions (logical OR)
data_cleaned['predicted_anomaly'] = np.maximum(data_cleaned['isolation_pred'], data_cleaned['lof_pred'])

# Calculate predicted risk score
data_cleaned['predicted_risk_score'] = data_cleaned['predicted_anomaly'] * data_cleaned['Base Pay']

# Output Risk Scoring for Predicted Anomalies
print("\nRisk Scoring for Predicted Anomalies:")
print(data_cleaned[['Base Pay', 'predicted_anomaly', 'predicted_risk_score']].head())

# Optional: Show summary statistics for predicted anomalies
predicted_anomalies = data_cleaned[data_cleaned['predicted_anomaly'] == 1]
print("\nSummary Statistics for Predicted Anomalies:")
print(predicted_anomalies['predicted_risk_score'].describe())

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# ==========================
# Autoencoder Model for Fraud Detection
# ==========================

# Select relevant features
features = ['Base Pay', 'Bureau', 'Office', 'Job Code', 'Position ID', 'Fiscal Year']

# Scale features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(data_cleaned[features])

# Train-test split
X_train, X_test = train_test_split(X_scaled, test_size=0.3, random_state=42)

# Define Autoencoder architecture
input_dim = X_train.shape[1]
input_layer = Input(shape=(input_dim,))
encoder = Dense(8, activation="relu")(input_layer)
encoder = Dense(4, activation="relu")(encoder)
decoder = Dense(8, activation="relu")(encoder)
decoder = Dense(input_dim, activation="sigmoid")(decoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer="adam", loss="mse")

# Train Autoencoder
history = autoencoder.fit(X_train, X_train,
                          epochs=20,
                          batch_size=32,
                          shuffle=True,
                          validation_data=(X_test, X_test),
                          verbose=1)

# Reconstruction error
reconstructions = autoencoder.predict(X_scaled)
mse = np.mean(np.power(X_scaled - reconstructions, 2), axis=1)

# Threshold for anomaly
threshold = np.percentile(mse, 95)  # top 5% as anomalies
data_cleaned['anomaly'] = (mse > threshold).astype(int)

# Risk scoring
data_cleaned['risk_score'] = data_cleaned['anomaly'] * data_cleaned['Base Pay']

# Print anomalies
print("\nDetected Anomalies (Autoencoder):")
print(data_cleaned[data_cleaned['anomaly'] == 1].head())

# Visualize anomaly distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=data_cleaned, x='anomaly')
plt.title('Anomaly Detection using Autoencoder')
plt.xlabel('Anomaly')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Normal', 'Anomaly'])
plt.show()

# Evaluation (if ground truth available)
if 'label' in data_cleaned.columns:
    print("\nClassification Report:")
    print(classification_report(data_cleaned['label'], data_cleaned['anomaly']))
    print("ROC-AUC Score:", roc_auc_score(data_cleaned['label'], data_cleaned['anomaly']))

# Clustering anomalies (KMeans)
from sklearn.cluster import KMeans
anomalies = data_cleaned[data_cleaned['anomaly'] == 1]
anomalies_features = anomalies[['Base Pay']]
kmeans = KMeans(n_clusters=3, random_state=42)
anomalies['cluster'] = kmeans.fit_predict(anomalies_features)

# Cluster visualization
plt.figure(figsize=(12, 6))
sns.scatterplot(data=anomalies, x='Base Pay', y='risk_score', hue='cluster', palette='viridis')
plt.title('Clustering of Detected Anomalies (Autoencoder)')
plt.xlabel('Base Pay')
plt.ylabel('Risk Score')
plt.legend(title='Cluster')
plt.show()
```
